```python
import os
import numpy as np
import cv2
import darknet
```

```python
def draw_rects(img, rects,path):
    """
    ?????????????
    :param img: 
    :param rects: 
    :return: 
    """
    
    for x, y, w, h in rects:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 00), 2)
        face = img
        text = "Cyclist"
        cv2.putText(img, text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), lineType=cv2.LINE_AA) 
    status = cv2.imwrite(path+'result.png', img)
    #cv2.imshow("Output",img)
    #cv2.waitKey()   
    #cv2.destroyAllWindows()
```

```python
files = os.listdir('./test/results')
for i in range(len(files)):
    xpo = []
    ypo = []
    cfiles = os.listdir("./test/results/"+files[i])
    if len(cfiles) > 1:
        f1 = open("./test/results/"+files[i]+'/'+"pr.txt", "r")    
        for m in f1:
            xpo.append(m.split()[0])
            ypo.append(m.split()[1])
        predictions = []
        no_obj = []    
        tfiles = os.listdir("./test/results/"+files[i]+'/')
        for j in tfiles:
            if "png" in j and "result" not in j:
                number = j.split('_')[4].split('.png')[0]
                output = darknet.performDetect(imagePath="./test/results/"+files[i]+'/'+j, thresh= 0.50, configPath = "./cfg/yolov3-tiny-obj.cfg", weightPath = "./backup/yolov3-tiny-obj_best.weights", metaPath= "./data/obj.data", showImage= True, makeImageOnly = False, initOnly= False)
                predictions.append([output,number])
            
        objects_per_region = []
        for k in range(len(predictions)):
            if len(predictions[k][0]['detections']) > 0:
                left = []
                top = []
                right = []
                bottom = []
                for l in range(len(predictions[k][0]['detections'])):
                    x = predictions[k][0]['detections'][l][2][0] - (predictions[k][0]['detections'][l][2][2]/2)
                    y = predictions[k][0]['detections'][l][2][1] - (predictions[k][0]['detections'][l][2][3]/2)
                    left.append(x)
                    top.append(y)
                    right.append(predictions[k][0]['detections'][l][2][2])
                    bottom.append(predictions[k][0]['detections'][l][2][3])
                    number = predictions[k][1]
                objects_per_region.append([left,top,right,bottom,number])
            else:
                no_obj.append(k)
        rects = []
        for n in range(len(objects_per_region)):
            if n not in no_obj:
                for p in range(len(objects_per_region[n][0])):
                    number = int(objects_per_region[n][4])
                    rects.append([int(objects_per_region[n][0][p])+int(np.round(float(xpo[number-1]))),int(objects_per_region[n][1][p])+int(np.round(float(ypo[number-1]))),int(objects_per_region[n][2][p]),int(objects_per_region[n][3][p])])
        im = cv2.imread("./test/imgs/"+files[i]+'.png')        
        draw_rects(im,rects,'./test/results/'+files[i]+'/')
```
