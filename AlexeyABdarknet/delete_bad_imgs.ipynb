```python
import os
```

```python
train = open("./data/train.txt","r")
test = open("./data/test.txt","r")
```

```python
for f in train:
    print f
    break
```

```python
ntrain = open("./data/new_train.txt",'w')
ptfiles = os.listdir("./data/obj/")
for f in train:
    if f.split("\n")[0].split("/")[2] in ptfiles:
        ntrain.write(f)
train.close()
ntrain.close()
```

```python
ntest = open("./data/new_test.txt",'w')
ptfiles = os.listdir("./data/obj/")
for f in test:
    if f.split("\n")[0].split("/")[2] in ptfiles:
        ntest.write(f)
test.close()
ntest.close()
```
