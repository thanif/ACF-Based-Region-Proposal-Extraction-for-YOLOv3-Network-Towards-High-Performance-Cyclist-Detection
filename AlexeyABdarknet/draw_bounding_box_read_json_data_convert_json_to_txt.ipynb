```python
src_path = "/media/talha/73fb317f-f61e-4b1f-b59e-5121d06d212b/Code(ACF)/data/Tsinghua-Daimler/new_test_set/posGt/"
names = os.listdir(src_path)
#dst_path = "/media/talha/73fb317f-f61e-4b1f-b59e-5121d06d212b/Code(ACF)/data/Tsinghua-Daimler/test/labels/"
imgs_categorized = 0 
easy = []
moderate = []
hard = []

test_path = os.listdir("/media/talha/efcaac58-26d5-464a-b9b8-0c0718c2011e/Tsinghua-Daimler Dataset/Dataset/Scripts/tsinghuaDaimlerScripts/Data/labelData/test/tsinghuaDaimlerDataset/")
valid_path = os.listdir("/media/talha/efcaac58-26d5-464a-b9b8-0c0718c2011e/Tsinghua-Daimler Dataset/Dataset/Scripts/tsinghuaDaimlerScripts/Data/labelData/valid/tsinghuaDaimlerDataset/")
    
not_counted = 0

for name in names:
        
    if name.split(".")[0]+"_labelData.json" in test_path:
        
        path = "/media/talha/efcaac58-26d5-464a-b9b8-0c0718c2011e/Tsinghua-Daimler Dataset/Dataset/Scripts/tsinghuaDaimlerScripts/Data/labelData/test/tsinghuaDaimlerDataset/"
        
    else:
        
        path = "/media/talha/efcaac58-26d5-464a-b9b8-0c0718c2011e/Tsinghua-Daimler Dataset/Dataset/Scripts/tsinghuaDaimlerScripts/Data/labelData/valid/tsinghuaDaimlerDataset/"
    
    with open(path+name.split(".")[0]+"_labelData.json", 'r') as f:
        distros_dict = json.load(f)
        
    br_name = name.split("_")
    
    fname = br_name[0]+"_"+br_name[1]+"_"+br_name[2]+"_"+br_name[3]+".txt"    
    
    category = {}
  
    for cyc in range(len(distros_dict["children"])):
        
        if len(distros_dict["children"][cyc]['tags']) == 0 and (distros_dict["children"][cyc]['maxcol'] - distros_dict["children"][cyc]['mincol'] > 60):
            
            category[name+'_'+str(cyc)] = 1
        
        elif len(distros_dict["children"][cyc]['tags']) > 0:
            
            
            if distros_dict["children"][cyc]['tags'][0].split(">")[0] == "occluded":
            
                if int(distros_dict["children"][cyc]['tags'][0].split(">")[1]) < 40 and (distros_dict["children"][cyc]['maxcol'] - distros_dict["children"][cyc]['mincol'] > 45):
                    
                    category[name+'_'+str(cyc)] = 2
                    
                elif int(distros_dict["children"][cyc]['tags'][0].split(">")[1]) >= 40 and int(distros_dict["children"][cyc]['tags'][0].split(">")[1]) < 80 and (distros_dict["children"][cyc]['maxcol'] - distros_dict["children"][cyc]['mincol'] > 30):
                    
                    category[name+'_'+str(cyc)] = 3
                    
                    
    c = category.values()
    m = {}
    m[1] = 0
    m[2] = 0
    m[3] = 0
    for i in c:
        
        m[i] += 1   
        
    #print m.values()        
    if max(m.values()) >0: 
        #print np.argmax(m.values())
        imgs_categorized +=1
        if np.argmax(m.values()) == 0:
            easy.append(name)
        elif np.argmax(m.values()) == 1:
            moderate.append(name)
        else:
            hard.append(name)
    else:
        
        not_counted+=1
    
    
            
    
```
