```python
import os
```

```python
files = os.listdir("./obj/")
```

```python
img_files = []
for f in files:
    if "jpg" in f:
        img_files.append(f)
```

```python
len(img_files)
```

```python
train = img_files[:int(len(img_files)*0.7)]
test = img_files[int(len(img_files)*0.7):]
len(train),len(test)
```

```python
file = open('train.txt','w') 
for f in train: 
    file.write('data/obj/'+f+'\n') 
file.close()     
```

```python
file = open('test.txt','w') 
for f in test: 
    file.write('data/obj/'+f+'\n') 
file.close()     
```
